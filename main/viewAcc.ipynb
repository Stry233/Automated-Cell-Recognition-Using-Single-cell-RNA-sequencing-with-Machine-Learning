# Prepareration for visualization
with variable controlled

```python
def getPlotDictKeepClassifier(inputList, classifier):
    """
    This function will return the dict that
    contain dimension reduction as key and 
    its acc as value.
    
    Parameters
    ----------
    inputList : list(string)
        list contains all info
        
    classifier: string
         the mode options for specific classifier part parsing in inputList
         check your classifer name in inputList to see the choice
         we always keep it the same, so if you want multiple classifier result,
         then call this for multiple times
    
    Returns
    -------
    dict:
        the dict that contain dimension reduction as key and its acc as value
    """      
    res = dict()
    for elementIndex in range(len(inputList)):
        clsName = rawData[elementIndex][1].split(": ")[-1] # name of classifier
        if (clsName != classifier): continue # since we set classifier unchange
        
        # set name of dimension reduction
        if ("+" in rawData[elementIndex][0]):  # CASE: combination with FIRST CHANGE
            firstName = rawData[elementIndex][0].split()[2].split("(")[0]
            secondName = "".join(rawData[elementIndex][0].split()[-2::])
            drName = firstName + secondName + " ({} Changing)".format(firstName)
        else: # single
            drName = rawData[elementIndex][0].split(": ")[-1].split("(")[0]
        acc = float(rawData[elementIndex][2].split(": ")[-1][:-1:]) # accuracy
        
        if (rawData[elementIndex][0].split("=")[-1].replace(")", "").isdigit()): # check if dimemsion is got
            dimension = float(rawData[elementIndex][0].split("=")[-1].replace(")", ""))
        elif (rawData[elementIndex][0].split("=")[-1].split(")")[0].isdigit()):
            dimension = float(rawData[elementIndex][0].split("=")[-1].split(")")[0])
        else:
            dimension = 2.0 # default
        if (drName not in res):
            res[drName] = [[dimension, acc]]
        else:
            res[drName].append([dimension, acc])
    return res
            
```

```python
def getPlotDictChangeClassifier(inputList):
    """
    This function will return the dict that
    contain parameter of classifier as key and 
    its acc as value.
    
    Parameters
    ----------
    inputList : list(string)
        list contains all info
    
    Returns
    -------
    dict:
        the dict that contain parameter of classifier as key and its acc as value
    """      
    res = []
    for elementIndex in range(len(inputList)):
        res.append([int(inputList[elementIndex][1].split("=")[-1].split(")")[0]), float(inputList[elementIndex][2].split(": ")[-1][:-1:])])
    return res
            
```

```python
# test

for typeNow in allClassifier:
    print("set classifier {} same, we have {}\n".format(typeNow, getPlotDictKeepClassifier(rawData, typeNow)))
```

# visualization

```python
ax = plt.subplot(1,1,1)
modelUsed = "GaussianNB"
plt.title("Accuracy result comparision: " + modelUsed)
ax.set_ylabel('acc(%)', fontdict={'size': 15})
ax.set_xlabel('dimension', fontdict={'size': 15})
accRes = getPlotDictKeepClassifier(rawData, modelUsed)
for dictKeys, dictVals in accRes.items():
    realKeys = dictKeys + " + " + modelUsed
    #plt.plot(max(list(x[0] for x in dictVals)),max(list(x[1] for x in dictVals)),'ko') 
    plt.plot(list(x[0] for x in dictVals if x[0] <= 80), list(x[1] for x in dictVals if x[0] <= 80), \
             label= realKeys, marker='o', markersize=3)
plt.gcf().set_size_inches(30, 10)
plt.legend()
fileName = "ACC_setClassifierSame_"+modelUsed+".jpg"
plt.savefig(fileName)
```
